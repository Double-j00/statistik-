```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("car_price_dataset.csv", sep=";")
```

```python
df
```

```python
display(df.head())
display(df.info())
```

```python
display(df.describe())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Price"], bins=30, kde=True)
plt.title("Prisfördelning av bilar")
plt.xlabel("Pris")
plt.ylabel("Antal bilar")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Mileage"], y=df["Price"], alpha=0.5)
plt.title("Samband mellan körsträcka och pris")
plt.xlabel("Körsträcka (km)")
plt.ylabel("Pris")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Fuel_Type"], y=df["Price"])
plt.title("Prisfördelning per bränsletyp")
plt.xlabel("Bränsletyp")
plt.ylabel("Pris")
plt.show()
```
